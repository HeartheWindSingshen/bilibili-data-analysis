# 1. 数据预处理 

首先导入数据，去掉数据中有缺失的部分和重复的部分。

# 3.1各个季度产出视频数量和对应的观看，投币，弹幕，评论等的数量关系。

我们可以看到的是，视频产出的数量基本与播放量，投币，弹幕，评论数量正类相关。
同时我们发现2019年第一季度的视频产出量比较低，后面逐季度上升，考虑到由于短视频2019年附近的开始火爆和B站最近几年开始越来越出圈。驻站的up主也越来越多。视频产量也就上去了。

```python
plt.figure(figsize=(15,8))
gindex=df.index
for i in gindex:
    k=list(df.at[i,'date'])
    t=0
    for j in range(len(k)):
        if k[j]!='/':
            t+=1
        else:
            if t==1:
                k[j-1]='0'+k[j-1]
            t=0
        if j==len(k)-1:
            k[j]=k[j]+":00"
    k="".join(k)
    df.at[i,'date']=datetime.datetime.strptime(k, '%m/%d/%y %H:%M:%S')
print(df["date"])
df=df.sort_values('date')
df1=df[df['date']<datetime.datetime(2019,4,1,0,0)]
df2=df[df['date']<datetime.datetime(2019,7,1,0,0)]
df2=df2[df2['date']>=datetime.datetime(2019,4,1,0,0)]
df3=df[df['date']<datetime.datetime(2019,10,1,0,0)]
df3=df3[df3['date']>=datetime.datetime(2019,7,1,0,0)]
df4=df[df['date']<datetime.datetime(2020,1,1,0,0)]
df4=df4[df4['date']>=datetime.datetime(2019,10,1,0,0)]
df5=df[df['date']>datetime.datetime(2020,1,1,0,0)]
d1=[len(df1),len(df2),len(df3),len(df4),len(df5)]
d2=[df1['view'].sum(),df2['view'].sum(),df3['view'].sum(),df4['view'].sum(),df5['view'].sum()]
d3=[df1['likes'].sum(),df2['likes'].sum(),df3['likes'].sum(),df4['likes'].sum(),df5['likes'].sum()]
d4=[df1['favorite'].sum(),df2['favorite'].sum(),df3['favorite'].sum(),df4['favorite'].sum(),df5['favorite'].sum()]
d5=[df1['coins'].sum(),df2['coins'].sum(),df3['coins'].sum(),df4['coins'].sum(),df5['coins'].sum()]
d6=[df1['share'].sum(),df2['share'].sum(),df3['share'].sum(),df4['share'].sum(),df5['share'].sum()]
d7=[df1['replay'].sum(),df2['replay'].sum(),df3['replay'].sum(),df4['replay'].sum(),df5['replay'].sum()]
d8=[df1['danmu'].sum(),df2['danmu'].sum(),df3['danmu'].sum(),df4['danmu'].sum(),df5['danmu'].sum()]
k=["2019年第一季度","2019年第二季度","2019年第三季度","2019年第四季度","2020年第一季度"]
adx1=plt.subplot2grid((8,2),(0,0),colspan=1,rowspan=1)
adx1.bar(k,d1,facecolor='red',edgecolor='white')
plt.title('各季度发布视频数量')
adx2=plt.subplot2grid((8,2),(0,1),colspan=1,rowspan=1)
adx2.bar(k,d2,facecolor='green',edgecolor='white')
plt.title('各季度的总观看量')
adx3=plt.subplot2grid((8,2),(2,0),colspan=1,rowspan=1)
adx3.bar(k,d3,facecolor='pink',edgecolor='white')
plt.title('各季度的点赞总量')
adx4=plt.subplot2grid((8,2),(2,1),colspan=1,rowspan=1)
adx4.bar(k,d4,facecolor='pink',edgecolor='white')
plt.title('各季度的收藏总量')
adx5=plt.subplot2grid((8,2),(4,0),colspan=1,rowspan=1)
adx5.bar(k,d5,facecolor='pink',edgecolor='white')
plt.title('各季度的硬币总量')
adx6=plt.subplot2grid((8,2),(4,1),colspan=1,rowspan=1)
adx6.bar(k,d6,facecolor='pink',edgecolor='white')
plt.title('各季度的分享总量')
adx7=plt.subplot2grid((8,2),(6,0),colspan=1,rowspan=1)
adx7.bar(k,d7,facecolor='pink',edgecolor='white')
plt.title('各季度的评论总量')
adx8=plt.subplot2grid((8,2),(6,1),colspan=1,rowspan=1)
adx8.bar(k,d8,facecolor='pink',edgecolor='white')
plt.title('各季度的弹幕总量')
```

# 4 我们想使用K-means聚类算法根据视频火热程度将视频分为四类（冷门，中等，热门，超火爆），而评测视频火热程度的参数选取了点赞，投币和评论数量来进行预测。最终画出三维的聚类图像。

首先我们先创建三维窗口界面，取出硬币，点赞，评论的所有数据。

```python
from sklearn.cluster import KMeans
from mpl_toolkits.mplot3d import Axes3D

X = np.array(list(zip(df['coins'].head(30000), df['likes'].head(30000),df['replay'].head(30000)))).reshape(30000, 3)
print(X)
```

我们设置不同类别的颜色，形状，并将三组数据存入到三个列表中用于之后的绘制。
由于导出页面限制，我们只展示前1000个数据

```python
colors = ['b', 'g', 'r', 'y']
markers = ['+', '+', 'x','x']
list11=[]
list22=[]
list33=[]
for i in df['replay'].head(30000).index:
    list11.append(df['replay'][i])
for i in df['coins'].head(30000).index:
    list22.append(df['coins'][i])
for i in df['likes'].head(30000).index:
    list33.append(df['likes'][i])
print(list11[0:1000])
print(list22[0:1000])
print(list33[0:1000])
```

```python
K = 4
kmeans_model = KMeans(n_clusters=K).fit(X)
print(list(enumerate(kmeans_model.labels_))[0:1000])
```

绘制三维图像

```python
fig2=plt.figure(figsize=(10,8))
ax2=Axes3D(fig2)
for i, l in enumerate(kmeans_model.labels_):
    ax2.scatter(list11[i],list22[i],list33[i],c=colors[l],marker=markers[l],s=30,alpha=0.5)
```

我们看到图像数据被分为四类，其中蓝色部分数量很多，其次是黄色部分，而绿色部分数量略显稀疏，红色部分数量极少。
相应的我们看到数量多的部分都是点赞，投币评论数量相对较少的，而数量少的部分则点赞，投币和评论数量都比较高，这与B站视频热门程度分布情况大致相同。

# 5 数据结论

通过以上的数据分析，我们对不同分区不同季度的视频数量分布有一定的了解，并且根据数据统计知晓了播放量，视频产出量，投币量点赞量等前十的视频或者up主，最容易被白嫖的视频和up主。也了解了2019年各个季度B站视频的火热程度，也发现B站在2019年第一季度随着短视频风口逐渐出圈。我们同时发现观看量，点赞量，投币量，评论量，收藏量虽然呈正向相关但是并没有密切的互相联系，相反有可能点赞数量多的视频评论数量很少，而评论多的视频点赞数量却很少。官方媒体在B站视频播放量极高，大大实现了了传播新闻和中国社会主义思想。对于社会热点时间观众的讨论量一般极高，评论数量非常多。
同时播放数量也因此很多。我们看到不同分区的视频由于不同内容的接受门槛高低的不同以及受众的范围，发现一些比较冷门且理解门槛高的视频很难出圈，也就是说很难有较高的视频播放量，仅限于圈内火爆。而一些大众较能接受且很容易理解，吸引人的视频种类比较容易热门，上线高。但不论怎么样，分区的建立绝不是以播放量火热程度为目的，各个分区的观众都有他们对这些分区独特的感情和热爱，各个分区的up主都有对自己所在分区的一份一份视频的奉献，每份认真努力都被观众所喜爱，所赞扬。正是在这种氛围下，B站的视频质量才没有滑坡，up主和观众们和谐共处，B站也在向未来迈着自己的步伐。
